```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create correlation matrix
correlation_matrix = df[['share_count', 'repin_count', 'comment_count']].corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True)
plt.title('User Engagement Metrics Relationships')
plt.show()
```

```python
# Load your data
df = pd.read_csv('TIF_Results.csv')

# 1. Simple Bar Chart - Compare engagement metrics
plt.figure(figsize=(10, 6))
plt.bar(['Shares', 'Repins', 'Comments'], 
        [df['share_count'].sum(), df['repin_count'].sum(), df['comment_count'].sum()],
        color=['red', 'blue', 'green'])
plt.title('Total Engagement: Shares vs Repins vs Comments')
plt.ylabel('Total Count')
plt.show()
```

* This chart compares the total number of shares, repins, and comments on all scraped images.
* Repins (blue) are the highest, showing users mostly save images to their boards.
* Shares (red) come next, meaning users also shared them outside.
* Comments (green) are very low, suggesting users rarely leave text feedback.
In short, Users engage more by saving and sharing than by commenting.

```python
# Simple pie chart
total_shares = df['share_count'].sum()
total_repins = df['repin_count'].sum() 
total_comments = df['comment_count'].sum()

plt.pie([total_shares, total_repins, total_comments], 
        labels=['Shares', 'Repins', 'Comments'],
        autopct='%1.1f%%')
plt.title('How Users Engage with Food Images')
plt.show()
```

* Repins (81.6%) are the most common — users mainly save the images.
* Shares (18.2%) are the second, meaning some users spread content outside.
* Comments (0.2%) are rare, showing users hardly leave written feedback.
